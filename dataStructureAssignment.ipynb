# 19 Why are dictionaries considered unordered, and how does that affect data retrieval?
answer> Dictionaries in Python (and many other programming languages) are considered unordered because they store data in a way that does not guarantee the order of elements. Specifically, dictionaries store data as key-value pairs, and while you can iterate over the keys or values in a dictionary, there is no inherent order to how these pairs are stored or retrieved. This characteristic is related to the internal structure of a dictionary, which is typically implemented as a hash table.
        Dictionaries are optimized for fast lookups, additions, and deletions. The focus on speed leads to the use of hash tables, where elements are placed at locations determined by their hash values, not in the order they were inserted.

# 20 Explain the difference between a list and a dictionary in terms of data retrieval.
answer> A list is an ordered collection of items, whereas a dictionary is an unordered data collection in a key: value pair. Elements from the list can be accessed using the index, while the elements of the dictionary can be accessed using keys.
